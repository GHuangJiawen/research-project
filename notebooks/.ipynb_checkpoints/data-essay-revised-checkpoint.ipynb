### Introduction
In the first half of the 19th century, the slavery system in the United States continued to develop with the expansion of its territory, and the contradiction between the North and the South regarding slavery also intensified increasingly. Against this historical backdrop, a large number of enslaved African Americans chose to seek freedom by running away (becoming "fugitive slaves"), but the scale and characteristics of these escape events lacked detailed personal records, and it has been difficult to conduct quantitative research on the demographic characteristics of the fugitive slave group in history. The Freedom on the Move (FOTM) project led by Cornell University has created a comprehensive database of fugitive slave advertisements in North America, gathering tens of thousands of fugitive slave notices from American newspapers and periodicals in the 19th century. These advertisements were usually published by slave owners, detailing the personal information of the escapees such as their names, genders, ages, and physical features. FOTM data provides valuable information for studying the struggles of the enslaved during the slavery period, enabling us to systematically examine the scale and patterns of the phenomenon of runaway slavery from a data perspective. 


Some existing studies have also explored similar directions. For example, Shaun Wallace (2017) analyzed over 5,000 advertisements for runaway slaves in Georgia and Maryland, published between 1790 and 1810. His research revealed a connection between gender, age, literacy, and rebellious behavior among those who escaped slavery.
He also emphasized that these advertisements did more than simply record escape events — they also preserved the details and subjectivity of enslaved people's resistance. Other studies have focused more on the analysis of the advertisement itself, while our research focuses on extensive statistics. Based on the data of the FOTM database, the research analyzes the basic demographic patterns and changes of the fugitive slaves in the United States in the 19th century. In other words, our research attempt to answer the following several specific questions through data:
1. Gender composition: What were the proportions of men and women in the group of runaway slaves?
2. Age distribution: What was the age distribution of the runaway slaves? Are there any significant differences in the age structure between male and female runaway slaves?
3. Geographical distribution: Which states or regions did these runaway slaves mainly flee from? Is there a regional concentration phenomenon?
4. Gender Influence of Slave Owners: Were there differences in the frequency of slave escapes experienced by slave owners of different genders and the gender composition of escapees?
5. Time Trend: How did the number of runaway slaves recorded each year change? Did these annual changes coincide with the major historical events during the expansion of slavery and the intensification of the conflict between the North and the South?

Through the statistical and visual analysis of these variables, this project attempts to answer the following core question: Against the historical background of the expansion of slavery and the intensification of North-South contradictions, what structural characteristics do the gender, age and geographical distribution of the group of runaway slaves present? How did these characteristics interact with the gender of the slave owners, institutional pressure and historical events?

### Methodology

Just as is said in the introduction, the dataset used in this project comes from the Freedom on the Move (FOTM) archive, a collaborative digital humanities project led by several universities and institutions. The original data began as fugitive slave ads printed in 18th- and 19th-century newspapers, typically placed by enslavers or jailers to recapture those who had escaped. These newspapers were preserved in archives, then digitized into PDF format, and later transcribed through a large-scale crowdsourcing initiative. After transcription, FOTM standardized much of the text into structured fields and published the resulting data in CSV format, accessible through their public-facing platform.

```python
import pandas as pd
import plotly.io as pio
import plotly.offline as pyo
pio.renderers.default = "jupyterlab"
pyo.init_notebook_mode(connected=True)

file_path = '../data/original/FOTM-Dataset-Full-Flattened.csv' 
df_1 = pd.read_csv(file_path, low_memory=False)
pd.set_option('display.max_columns', None)  
pd.set_option('display.expand_frame_repr', False)  

df_1.head() 
```

Although some fields in this dataset have been structured, there are still many missing values, inconsistent formats and outliers. In the data cleaning process of my project, I first conducted a systematic data cleaning on it. By standardizing using Python's Pandas library, capitalizing the first letter of all fields and replacing missing fields or null values with "Not_Provided". In terms of field filtering, I retained all columns with analytical value or historical significance., including enslaver_gender, publication_year, approximate_age and other variables used to reveal demographic and institutional characteristics. 


Meanwhile, I also retained the transcribed_text field. Although this project did not conduct text analysis on the full advertisement text, I think this is one of the most distinctive parts of this dataset, reflecting the original sound of historical records. Retaining it is a respect for the humanistic value of the data and also leaves room for possible qualitative research in the future. Some irrelevant fields (such as names, redundant columns or extremely sparse free text fields) are deleted to simplify the data structure and improve the analysis efficiency. After that, I found that in the approximate_age field, there were a lot of mixed numeric and literal values. I extracted the values, converted vague expressions like "about 25" and "estimated 30" to integers, and removed unreasonable records of those under 5 years or over 100 years old. To ensure the smoothness of subsequent analysis. Overall, during the data cleaning process, I strive to maintain the clarity of the analysis while respecting the complexity and historical context of the data.


It is worth noting that there are many missing values in some columns, and age is one of them. This will directly lead to our subsequent analysis being not very comprehensive, which will be mentioned in the subsequent analysis.

```python
def clean_cell(x):
    if pd.isnull(x) or str(x).strip().lower() in ['nan', 'none', '', 'NOT_PROVIDED']:
        return 'Not_Provided'
    else:
        return str(x).strip().title()
df_cleaned = df_1.applymap(clean_cell)
columns_to_keep = [
    'enslaved_person_id',
    'gender',
    'approximate_age',
    'skills',
    'literacy',
    'enslaved_person_name_fullname 1',
    'publication_date',
    'advertisement_id',
    'advertisement_transcription',
    'advertisement_language',
    'newspaper_location_state',
    'runaway_location_state 1',
    'enslaver_gender 1'
]
df_final = df_cleaned[columns_to_keep].copy()
df_final['publication_year'] = pd.to_datetime(df_final['publication_date'], errors='coerce').dt.year
df_final = df_final.drop(columns=['publication_date'])
df_final.to_csv('../data/processed/1-FOTM-Dataset-Full-Flattened.csv', index=False)
```

```python
import pandas as pd
df_raw = pd.read_csv('../data/processed/1-FOTM-Dataset-Full-Flattened.csv', low_memory=False)
def clean_age_preserve_not_provided(value):
    if str(value).strip().lower() == 'not_provided':
        return 'Not_Provided'
    try:
        digits = ''.join([ch for ch in str(value) if ch.isdigit()])
        age = int(digits) if digits else None
        if age is not None and 5 <= age <= 100:
            return str(age)  
        else:
            return 'Not_Provided'
    except:
        return 'Not_Provided'
df_raw['approximate_age'] = df_raw['approximate_age'].apply(clean_age_preserve_not_provided)
df_raw.to_csv('../data/processed/cleaned-FOTM-Dataset-Full-Flattened.csv', index=False)
df_raw.sample(5)
```

### Analysis

#### **Year**

By statistically analyzing the changing trends of the number of advertisements in different periods, we can identify the key historical nodes of slavery and observe when resistance behaviors concentrated and erupted and when they tended to decline.

First of all, it is worth noting that from 1828, the number of advertisements for runaway slaves began to increase significantly. This trend occurred before the Nat Turner's Rebellion in 1831 and might reflect the accumulation of tensions within the slave society. The increase in the act of runaway slaves not only implies the direct resistance of individuals against the oppressive system, but can also be regarded as a precursor to the brewing process of larger-scale slave uprisings. Slave owners strengthened their manhunt through newspapers, which was also a means of enhanced control they adopted when trust in the system began to waver. 


Entering the 1850s, there was a second surge in the number of advertisements, and the key legal turning point behind it was the enactment of the "Fugitive Slave Act of 1850". This act forced the northern states to assist the southern slave owners in hunting down runaway slaves and imposed severe penalties on those who did so, greatly expanding the legal dissemination range of runaway slave advertisements. The outbreak of the "Bleeding Kansas" incident in 1854 marked that the controversy over slavery had expanded into a national conflict. Under such a political atmosphere, the number of runaway slaves rose simultaneously with the volume of advertisements released, reflecting the high alertness of slave owners to the risk of institutional collapse.


The chart shows a significant decline in the number of advertisements around 1860. This change was kind of related to Lincoln's election as president, the successive secession of the southern states from the Union. Against the backdrop of fundamental questioning of the legitimacy of the system and the country falling into a crisis of division, the executive power and public opinion control of the slavery system weakened significantly. The decline in the number of advertisements for runaway slaves may reflect both the problems brought about by the disintegration of the system and the gradual loss of the ability of slave owners to control "property" on the verge of disorder.


When analyzing these year data of fugitive slave advertisements, we need to note that the rise of the cotton economy in the 1830s had a significant impact on the expansion of slavery. This period marked the explosive growth of the plantation economy in the South and might have led to changes in the number of fugitive slave advertisements. However, we should interpret the individual peaks in these data with caution, especially when the number of advertisements in certain years is much higher than that in others. Such peaks may not reflect the actual surge in flight activities. In some years, a large number of new newspapers may have begun to be published, or the original newspapers may have just been digitized, resulting in a sharp increase in the number of recorded advertisements. Therefore, although we can highlight these peaks, they may still reflect the irregularity of the data rather than a true runaway trend. When interpreting, it is necessary to remain vigilant about these potential deviations.

#### **Gender of the runaway slaves**
When analyzing the historical data of advertisements for runaway slaves, gender is an important dimension that cannot be ignored. The gender of the enslaved not only affects their division of labor and social roles within the slavery system, but may also profoundly influence the possibility of their escape. 

```python
import plotly.express as px
df_s = pd.read_csv('C:/Users/GAVIN/Desktop/research-project/data/processed/cleaned-FOTM-Dataset-Full-Flattened.csv')
df_gender_valid = df_s[df_s['gender'] != 'Not_Provided']
fig = px.pie(
    df_gender_valid,
    names='gender',
    title='Gender Distribution of Runaway Enslaved Persons (Excluding Not_Provided)',
    hole=0.4  
)
fig.show()
```

This chart shows the gender distribution of individuals enslaved in the runaway slave advertisement, including only records with specific gender information. Data shows that 78.5% of the fugitives are male, while only 21.5% are female. This significant imbalance indicates that enslaved men recorded escape attempts more frequently, either because they had more chances to escape or because the slave owners regarded them as more "valuable property" because their role was manual labor. Karen Cook Bell (2023) pointed out that, in contrast, enslaved women usually work in more secluded family Settings, such as at home, which makes escape more dangerous and easier to be detected, especially for those women who attempt to escape with their children. And Bell also said that escape advertisements and historical memories are disproportionately concentrated on men, thereby marginalizing women's stories, who also resist through escape and other means. The underrepresentation of women in the data records does not necessarily reflect a lower level of resistance, it may also be the product of historical silence and gender bias in the documents.

### **Age**
The age of the enslaved not only affects their labor role and living conditions in the slavery system, but also directly relates to the possibility of their escape. Analyzing the age distribution recorded in the advertisements of runaway slaves helps us understand which groups are most likely to adopt resistance behaviors and which ones may be under stricter control or in a more vulnerable position.

```python
df_age_valid = df_s[df_s['approximate_age'].apply(lambda x: str(x).isdigit())].copy()
df_age_valid['approximate_age'] = df_age_valid['approximate_age'].astype(int)
fig = px.histogram(
    df_age_valid,
    x='approximate_age',
    nbins=20,
    title='Age Distribution of Runaway Enslaved Persons',
    labels={'approximate_age': 'Approximate Age'},
    opacity=0.75
)
fig.update_layout(
    xaxis_title='Age',
    yaxis_title='Number of Ads',
    bargap=0.1
)
fig.show()
```

It can be seen from the chart that the ages of the enslaved are mainly concentrated between 20 and 40 years old, especially the age group of 21 to 30 years old is the most concentrated age group for fleeing. This trend indicates that the young and middle-aged years may be the stage when fleeing behavior is most frequent. In contrast, the number of runaway slaves aged 5 to 15 and over 60 was significantly smaller. This age distribution might be because young and middle-aged enslaved individuals usually undertake more arduous labor, are under greater pressure and surveillance, and also possess better physical strength, energy and the will to escape. Children or the elderly have a relatively lower chance of escaping due to reasons such as physical ability or family ties. Slave owners might be more inclined to invest in "productive" enslaved individuals and therefore would advertise more frequently for this age group

```python
# Count records with valid numeric ages
count_valid = df_s['approximate_age'].apply(lambda x: str(x).isdigit()).sum()

# Count records with missing or non-numeric ages
count_missing = len(df_s) - count_valid

# Prepare data for the pie chart
age_data = pd.DataFrame({
    'Age Status': ['Provided', 'Not Provided'],
    'Count': [count_valid, count_missing]
})

# Create the pie chart
fig = px.pie(
    age_data,
    names='Age Status',
    values='Count',
    title='Percentage of Records with and without Approximate Age',
    hole=0.4  # Set to 0.4 for a donut-style chart; use 0 for full pie
)

# Show the chart
fig.show()
```

We can see that a huge number of records are marked as "not provided". This absence not only affects our restoration of individual life trajectories, but also directly limits the representativeness and accuracy of the analysis of age distribution I made in the previous part. In other words, we need to understand that the age data that has been successfully quantified can only represent the part of people recorded in history, rather than the true structure of the entire group of runaway slaves.

#### **Male and Female Runaways by Age Group**
The intersection of gender and age is a key clue to understanding the living conditions of the enslaved and the possibility of resistance. Under the slavery system, the division of labor was highly dependent on gender and age structure. Different working environments also meant different possibilities of escape.

```python
df_age_gender = df_s[
    df_s['approximate_age'].apply(lambda x: str(x).isdigit()) &
    (df_s['gender'] != 'Not_Provided')
].copy()
df_age_gender['approximate_age'] = df_age_gender['approximate_age'].astype(int)
bins = [0, 10, 20, 30, 40, 50, 60, 100]
labels = ['0–10', '11–20', '21–30', '31–40', '41–50', '51–60', '61+']
df_age_gender['age_group'] = pd.cut(df_age_gender['approximate_age'], bins=bins, labels=labels, right=True)
age_gender_counts = df_age_gender.groupby(['age_group', 'gender']).size().reset_index(name='count')
fig = px.line(
    age_gender_counts,
    x='age_group',
    y='count',
    color='gender',
    markers=True,  
    text='count',
    title='Gender Distribution Across Age Groups of Runaway Enslaved Persons (Line Chart)',
    labels={'age_group': 'Age Group', 'count': 'Number of Ads'}
)
fig.update_traces(textposition='top center')
fig.update_layout(
    xaxis_title='Age Group',
    yaxis_title='Number of Ads',
    uniformtext_minsize=8,
    uniformtext_mode='hide'
)

fig.show()



```

Among all age groups, the number of male runaway slaves is generally higher than that of female slaves, especially in the age group of 21 to 30, where the number of men leads significantly. This indicates that young and middle-aged men are the main bearers of the fleeing behavior, which may be related to their more frequent engagement in outdoor labor, exposure to the external environment, and higher physical ability. However, the proportion of female runaway slaves among all age groups is relatively low, especially in middle and old age when there are almost no runaway records. This may reflect that they are subject to stronger spatial restrictions or constraints from family roles.


And since this analysis is based on previous age data, and most of the data is Not_Provided, the limitations are also obvious. However, we can see the big picture from the small details.

#### **Gender of the Enslavers**
The research holds that the gender composition of slave owners provides another important perspective for observing the power structure of the slave system. The gender of slave owners not only affects the language style and focus of advertisements, but may also have potential correlations with variables such as the gender of enslaved people and escape patterns.

```python
df_enslaver = df_s[df_s['enslaver_gender 1'] != 'Not_Provided'].copy()
enslaver_counts = df_enslaver['enslaver_gender 1'].value_counts().reset_index()
enslaver_counts.columns = ['Enslaver Gender', 'Count']
fig = px.pie(
    enslaver_counts,
    names='Enslaver Gender',
    values='Count',
    title='Gender Distribution of Enslavers (Based on Ads)',
    hole=0.3  
)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

The vast majority of advertisements for runaway slaves were posted by male slave owners (91.6%), and the proportion of female slave owners was significantly lower (7.36%). This distribution reflects the dominant position of men in property rights, social status and the operation of the system under the slave system. Since men are more often regarded as the "heads of households" in the legal and economic systems, they more frequently dominate the matters of wanted fugitives and property protection. Although women also have enslaved individuals in some family or inheritance contexts, their "voices" in public Spaces are relatively weakened, which is reflected in the relatively marginal number of advertisements for runaway slaves.

```python
# Filter valid gender data for both enslaver and enslaved
df_pair = df_s[
    (df_s['enslaver_gender 1'] != 'Not_Provided') &
    (df_s['gender'] != 'Not_Provided')
].copy()

# Count ads by enslaver gender and enslaved person gender
enslaver_slave_gender = df_pair.groupby(['enslaver_gender 1', 'gender']).size().reset_index(name='count')

# Calculate percentage within each enslaver gender group
enslaver_totals = enslaver_slave_gender.groupby('enslaver_gender 1')['count'].transform('sum')
enslaver_slave_gender['percentage'] = enslaver_slave_gender['count'] / enslaver_totals * 100

# Add label for display
enslaver_slave_gender['label'] = enslaver_slave_gender.apply(
    lambda row: f"{row['enslaver_gender 1']} → {row['gender']}<br>{row['percentage']:.1f}%",
    axis=1
)

# Plot 100% stacked bar chart
fig = px.bar(
    enslaver_slave_gender,
    x='enslaver_gender 1',
    y='percentage',
    color='gender',
    text='label',
    title='Percentage of Enslaved Person Gender by Enslaver Gender',
    labels={
        'enslaver_gender 1': 'Enslaver Gender',
        'gender': 'Enslaved Person Gender',
        'percentage': 'Percentage of Ads'
    },
    barmode='stack'
)

# Update visual layout
fig.update_traces(textposition='inside')
fig.update_layout(
    yaxis_title='Percentage of Ads',
    xaxis_title='Enslaver Gender',
    yaxis_ticksuffix='%',
    uniformtext_minsize=8,
    uniformtext_mode='hide'
)

fig.show()

```

A remarkable trend can be observed in the advertisements for runaway slaves: Most of the advertisements released by male slave owners were targeted at runaway male slaves, while in the advertisements released by female slave owners, the ratio of male to female runaway slaves was closer or even tended to be balanced. This phenomenon may reflect the management bias of slave owners of different genders in the division of slave roles: Male slave owners were more likely to manage male slaves engaged in heavy manual labor, while female slave owners were more likely to directly manage indoor laborers or domestic slaves more frequently, many of whom were women.

This not only reveals how the gender of the slave owner affects the subjects recorded in the advertisements of runaway slaves, but also refutes the prejudice within slavery against female slave owners. As Stephanie E. Jones-Rogers (2019) pointed out in her research, female slave owners were not "subordinate" or "passive" in accepting slavery as described in traditional narratives. Instead, they actively participated in it and even demonstrated comparable capabilities to men in slavery management and slave market activities.

But what is curious is whether the states with a larger slave population have more advertisements? For the analysis, I sought out the 1840 census data of Louisiana and Virginia (linked in the bibliography), and found the total slave population of each state :LA 352,411, VA: 1,239,797 for comparison.

```python
import matplotlib.pyplot as plt

states = ['Virginia', 'Louisiana']
ads = [413, 8740]
enslaved_population = [1239797, 352411]

ads_per_1000 = [ads[i] / enslaved_population[i] * 1000 for i in range(2)]

fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(states, ads_per_1000, color=['#1f77b4', '#ff7f0e'])

for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{yval:.2f}', ha='center', va='bottom', fontsize=10)

ax.set_title('Runaway Slave Ads per 1,000 Enslaved People', fontsize=14)
ax.set_ylabel('Ads per 1,000 Enslaved People')
ax.set_ylim(0, max(ads_per_1000) + 5)

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

Before reading, I think everyone needs to understand that my analysis, using the total number of slaves counted in one year with the total number of slave advertisements over several decades is a bit inaccurate. But I think the most important thing is to analyze the overall trend.


Virginia had 1.24 million enslaved people in 1840, but there were only 413 advertisements for runaway slaves. After standardization, there were only 0.33 advertisements for every 1,000 slaves. By contrast, the slave population in Louisiana was only about 350,000, but it had 8,740 advertisements, that is, approximately 24.81 advertisements per 1,000 slaves. This huge difference indicates that the number of original advertisements cannot directly reflect the frequency of escape or "activity level", but may be affected by multiple factors such as the recording mechanism, publication frequency, law enforcement system, and the integrity of advertisement preservation. This is also what my research hopes everyone can understand.

### Conclusion


This study holds that the advertisements for runaway slaves reveal a highly structured slavery system: men, young and middle-aged people, and enslaved individuals in southern states are more likely to become the "escapees" visible in the advertisements, and the gender and geographical location of the slave owners also have significant influences. The advertisement of runaway slaves not only records the individual's rebellious behavior, but also reflects a structured slave society where all the path of runaway slaves are traceable.

More broadly, this dataset is located in the core cultural context of slavery in the United States, an era when slaves were managed, wanted and commercialized as property. Although these advertisements were written by slave owners with the original intention of capturing "escaped property", today, they have instead become evidence of the identity, resistance and expression of the enslaved. The use of such data requires sensitivity: We must be aware that the data source itself is the product of the oppressive system. We cannot consume its content from a neutral perspective, but should reveal the power structure, silent scars and the continuation of historical inequality behind it with a reflective attitude.
